## Ranking of attributes(RII)

```python
#a=frequency_ex_ques.apply(lambda x: x is not frequency_ex_ques['Respondent Number'])
```

```python
ranking_of_attributes=pd.DataFrame(columns=['sum_of_weights','RII','Rank'])
a=frequency_ex_ques.drop(['Respondent Number'], axis=1)
ranking_of_attributes['sum_of_weights']=a.sum()
ranking_of_attributes['RII']=(a.sum())/((a.max())*len(a))
ranking_of_attributes=ranking_of_attributes.sort_values(by='RII',ascending=False)
ranking_of_attributes.index.name='Attributes'
ranking_of_attributes['Rank']=range(1,1+len(ranking_of_attributes))
ranking_of_attributes
```
