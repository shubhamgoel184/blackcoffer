## Ranking of attributes(RII)

```python
#a=frequency_ex_ques.apply(lambda x: x is not frequency_ex_ques['Respondent Number'])
```

```python
ranking_of_attributes=pd.DataFrame(columns=['sum_of_weights','RII','Rank'])
a=frequency_ex_ques.drop(['Respondent Number'], axis=1)
ranking_of_attributes['sum_of_weights']=a.sum()
ranking_of_attributes['RII']=(a.sum())/((a.max())*len(a))
ranking_of_attributes=ranking_of_attributes.sort_values(by='RII',ascending=False)
ranking_of_attributes.index.name='Attributes'
ranking_of_attributes['Rank']=range(1,1+len(ranking_of_attributes))
ranking_of_attributes
```

```python
ranking_of_attributes2=pd.DataFrame(columns=['sum_of_weights','RII','Rank'])
b=severity_ex_ques
#b=severity_ex_ques.drop(['Respondent Number'], axis=1)
ranking_of_attributes2['sum_of_weights']=b.sum()
ranking_of_attributes2['RII']=(b.sum())/((b.max())*len(b))
ranking_of_attributes2=ranking_of_attributes2.sort_values(by='RII',ascending=False)
ranking_of_attributes2.index.name='Attributes'
ranking_of_attributes2=ranking_of_attributes2.drop(ranking_of_attributes2.index[43])
ranking_of_attributes2['Rank']=range(1,1+len(ranking_of_attributes2))
ranking_of_attributes2
```
